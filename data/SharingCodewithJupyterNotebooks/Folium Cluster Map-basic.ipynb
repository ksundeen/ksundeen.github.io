```python
import numpy as np
import os
import folium
from folium.plugins import MarkerCluster

# Create sample points
size = 100
lons = np.random.randint(-180, 180, size=size)
lats = np.random.randint(-90, 90, size=size)

# Get list of lats/longs to feed the popups
locations = list(zip(lats, lons))
print(locations)
popups = ['lon:{}<br>lat:{}'.format(lon, lat) for (lat, lon) in locations]

m = folium.Map(
    location=[np.mean(lats), np.mean(lons)],
    tiles='Cartodb Positron',
    zoom_start=1
)

# Create new Leaflet MarkerClusterGroup from js lib
marker_cluster = MarkerCluster(
    name='1000 clustered icons',
    overlay=True,
    control=True,
    icon_create_function=None
)

# Add all locations & assign popups to the MarkerClusterGroup
for k in range(size):
    location = lats[k], lons[k]
    marker = folium.Marker(location=location)
    popup = 'lon:{}<br>lat:{}'.format(location[1], location[0])
    folium.Popup(popup).add_to(marker)
    marker_cluster.add_child(marker)

marker_cluster.add_to(m)

folium.LayerControl().add_to(m)
```

```python
# Save output to html file
m.save(outfile='testcluster.html')
m
```
